```python
from transformers import (
    AutoTokenizer,
    AutoModelForMaskedLM,
    AutoModelForSeq2SeqLM,
    pipeline,
)

# Initialize models
models = {
    "BERT": "bert-base-uncased",
    "RoBERTa": "roberta-base",
    "T5": "t5-small",
}

def extract_keywords_with_bert(job_description):
    """Extract keywords using BERT."""
    tokenizer = AutoTokenizer.from_pretrained(models["BERT"])
    model = AutoModelForMaskedLM.from_pretrained(models["BERT"])
    mask_pipeline = pipeline("fill-mask", model=model, tokenizer=tokenizer)

    # Insert a mask token at various positions for better context
    masked_text = job_description[:500] + " [MASK]"  # Trims long text
    try:
        results = mask_pipeline(masked_text)
    except:
        return ["Error during extraction with BERT"]

    # Extract single or two-word results
    keywords = [result["token_str"].strip() for result in results if len(result["token_str"].split()) <= 2]
    return keywords

def extract_keywords_with_roberta(job_description):
    """Extract keywords using RoBERTa."""
    tokenizer = AutoTokenizer.from_pretrained(models["RoBERTa"])
    model = AutoModelForMaskedLM.from_pretrained(models["RoBERTa"])
    mask_pipeline = pipeline("fill-mask", model=model, tokenizer=tokenizer)

    # Use the <mask> token dynamically in context
    masked_text = f"{job_description[:500]} <mask>"
    try:
        results = mask_pipeline(masked_text)
    except:
        return ["Error during extraction with RoBERTa"]

    # Extract single or two-word results
    keywords = [result["token_str"].strip() for result in results if len(result["token_str"].split()) <= 2]
    return keywords

def extract_keywords_with_t5(job_description):
    """Extract keywords using T5."""
    tokenizer = AutoTokenizer.from_pretrained(models["T5"])
    model = AutoModelForSeq2SeqLM.from_pretrained(models["T5"])
    t5_pipeline = pipeline("text2text-generation", model=model, tokenizer=tokenizer)

    # Prompt T5 with a specific task
    prompt = f"Extract key technologies, tools, and skills from the following job description: {job_description}"
    try:
        results = t5_pipeline(prompt, max_length=100, num_return_sequences=1)
    except:
        return ["Error during extraction with T5"]

    # Split results into keywords (limit to 1-2 words)
    keywords = results[0]["generated_text"].split(", ")
    return [kw.strip() for kw in keywords if len(kw.split()) <= 2]

# Dynamic job description input
job_description = input("Paste the job description: ")

# Extract keywords with each model
results = {}
print("\nExtracting keywords...")
results["BERT"] = extract_keywords_with_bert(job_description)
results["RoBERTa"] = extract_keywords_with_roberta(job_description)
results["T5"] = extract_keywords_with_t5(job_description)

# Display results
for model_name, keywords in results.items():
    print(f"\n{model_name} Keywords:")
    print(keywords)
```

```python
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM
from keybert import KeyBERT
import re

# Initialize KeyBERT for keyword extraction
kw_model = KeyBERT()

# Hugging Face Fine-tuned Model
t5_model = "t5-small"
t5_tokenizer = AutoTokenizer.from_pretrained(t5_model)
t5 = AutoModelForSeq2SeqLM.from_pretrained(t5_model)

def clean_keywords(keywords):
    """Ensure keywords are 1-2 words long and remove duplicates."""
    cleaned = []
    for kw in keywords:
        kw = kw.strip().lower()
        if 1 <= len(kw.split()) <= 2:  # Restrict to 1-2 words
            cleaned.append(kw)
    return list(set(cleaned))  # Remove duplicates

def extract_with_t5(job_description, section):
    """Extract relevant content for a specific section using T5."""
    try:
        t5_pipeline = pipeline("text2text-generation", model=t5, tokenizer=t5_tokenizer)
        prompt = f"Extract {section} in 1-2 words from the following job description: {job_description}"
        result = t5_pipeline(prompt, max_length=100, num_return_sequences=1)
        extracted = result[0]["generated_text"].split(", ")
        return clean_keywords(extracted)
    except Exception as e:
        return [f"T5 Error: {str(e)}"]

def extract_with_keybert(job_description):
    """Extract keywords dynamically using KeyBERT."""
    try:
        keywords = kw_model.extract_keywords(
            job_description, keyphrase_ngram_range=(1, 2), stop_words="english", top_n=15
        )
        return clean_keywords([kw[0] for kw in keywords])
    except Exception as e:
        return [f"KeyBERT Error: {str(e)}"]

# Dynamic job description input
job_description = input("Paste the job description: ")

# Extract Core Responsibilities and Required Skills using T5
core_responsibilities = extract_with_t5(job_description, "core responsibilities")
required_skills = extract_with_t5(job_description, "required skills")

# Extract general keywords using KeyBERT
keybert_keywords = extract_with_keybert(job_description)

# Generate Output
result = {
    "Core Responsibilities": core_responsibilities,
    "Required Skills": required_skills,
    "Keywords": keybert_keywords
}

# Print Results
print("\nExtracted Results:")
for key, value in result.items():
    print(f"{key}: {value}")
```

```python
import spacy
from keybert import KeyBERT

# Load SpaCy for text parsing
nlp = spacy.load("en_core_web_sm")

# Initialize KeyBERT for keyword extraction
kw_model = KeyBERT()

def extract_keywords_and_sections(jd):
    doc = nlp(jd)
    sections = {"Responsibilities": [], "Required Skills": [], "General": []}
    current_section = "General"  # Default section

    # Simple rule: Split based on common headings
    for sent in doc.sents:
        text = sent.text.strip().lower()
        if "responsibilities" in text:
            current_section = "Responsibilities"
        elif "requirements" in text or "skills" in text:
            current_section = "Required Skills"
        else:
            # Add sentences to the current section
            sections[current_section].append(sent.text.strip())

    # Keyword extraction using KeyBERT
    for section in sections:
        text = " ".join(sections[section])
        keywords = kw_model.extract_keywords(text, keyphrase_ngram_range=(1, 2), stop_words="english", top_n=20)
        sections[section] = [kw[0] for kw in keywords]

    return sections

# Example JD
job_description = input("Paste the job description: ")

# Extract keywords and sections
result = extract_keywords_and_sections(job_description)

# Display results
print("\nExtracted Keywords:")
for section, keywords in result.items():
    print(f"{section}: {keywords}")
```

```python
import requests
from requests.auth import HTTPBasicAuth
import xml.etree.ElementTree as ET
import pandas as pd

# O*NET API credentials
USERNAME = "sjsu1"  # Replace with your username
PASSWORD = "8776bcx"  # Replace with your password

# Base URL for O*NET API
BASE_URL = "https://services.onetcenter.org/ws/online"

# Function to fetch job data by keyword
def fetch_jobs_by_keyword(keyword, max_results=50):
    """
    Fetches job titles, SOC codes, and details URL from O*NET API for a specific keyword.

    Args:
        keyword (str): Keyword to search for (e.g., 'engineer', 'teacher').
        max_results (int): Maximum number of results to fetch.

    Returns:
        list: List of dictionaries containing job titles, SOC codes, and details URL.
    """
    url = f"{BASE_URL}/search?keyword={keyword}&start=1&end={max_results}"
    response = requests.get(url, auth=HTTPBasicAuth(USERNAME, PASSWORD))

    if response.status_code != 200:
        print(f"Error fetching data for keyword '{keyword}': {response.status_code}")
        return []

    # Parse the XML response
    try:
        root = ET.fromstring(response.content)
        jobs = []
        for occupation in root.findall("occupation"):
            title = occupation.find("title").text
            soc_code = occupation.find("code").text
            details_url = occupation.get("href")
            jobs.append({"Job Title": title, "SOC Code": soc_code, "Details URL": details_url})
        return jobs
    except Exception as e:
        print(f"Error parsing XML for keyword '{keyword}': {e}")
        return []

# Function to fetch job details using Details URL
def fetch_job_details(details_url):
    """
    Fetches job description and keywords for a specific job from O*NET API.

    Args:
        details_url (str): URL for detailed job information.

    Returns:
        dict: A dictionary containing the job description and keywords.
    """
    response = requests.get(details_url, auth=HTTPBasicAuth(USERNAME, PASSWORD))
    if response.status_code != 200:
        print(f"Error fetching job details: {response.status_code}")
        return {"Description": "N/A", "Keywords": "N/A"}

    # Parse the XML response
    try:
        root = ET.fromstring(response.content)
        description = root.find("description").text if root.find("description") is not None else "N/A"

        # Extract keywords from the XML
        keywords_elements = root.findall(".//keyword")  # Adjust this path if necessary
        keywords = [kw.text for kw in keywords_elements if kw.text]

        if not keywords:
            print(f"No keywords found for URL: {details_url}")

        return {"Description": description, "Keywords": ", ".join(keywords) if keywords else "N/A"}
    except Exception as e:
        print(f"Error parsing job details for URL '{details_url}': {e}")
        return {"Description": "N/A", "Keywords": "N/A"}

# List of technical and non-technical job categories to search
job_categories = ["software engineer", "data analyst", "teacher", "nurse"]

# Fetch and combine job data
all_jobs = []
for category in job_categories:
    print(f"Fetching jobs for category: {category}")
    job_list = fetch_jobs_by_keyword(category, max_results=20)  # Adjust max_results as needed

    # Fetch additional details for each job
    for job in job_list:
        details = fetch_job_details(job["Details URL"])
        job.update(details)
        all_jobs.append(job)

# Save the combined data to a DataFrame
df = pd.DataFrame(all_jobs)

# Save data to a CSV file
csv_file = "onet_job_data_with_keywords.csv"
df.to_csv(csv_file, index=False)

print(f"Data saved to {csv_file}")
```
